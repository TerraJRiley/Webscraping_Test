## The Actual Cleaning Process

```python
df_scraped["Name"] = df_scraped["Name"].fillna("Scraper Failure Mystery")
df_cleaned = df_scraped[~df_scraped['Name'].str.startswith("Scrape")]
```

```python
# List of column names to merge
column_names_to_merge = [col for col in df_cleaned.columns if col.startswith('Details About ')]

# Merging the specified columns into one
df_cleaned['Details About Profile (About)'] = df_cleaned[column_names_to_merge].apply(
    lambda row: ' '.join(row.values.astype(str)).replace("nan", "").strip(" "), axis=1)

# Optionally, drop the original 'Column_' columns
df_cleaned = df_cleaned.drop(column_names_to_merge, axis=1)
```

```python
df_cleaned["First Name"] = df_cleaned["Name"].map(lambda x: x.split(" ")[0])
```

```python
df_cleaned["Last Name"] = df_cleaned["Name"].map(lambda x: x.strip(" ").split(" ")[-1])
```

```python
df_cleaned = df_cleaned.rename({
    "Work and education"       : "Work & Education (About)",
    "Places lived"             : "Places Lived (About)",
    "Contact and basic info"   : "Contact and basic info (About)",
    "Family and relationships" :"Family & Relationships (About)",
    "Life events"              : "Life Events (About)"}, axis = 1)
```
